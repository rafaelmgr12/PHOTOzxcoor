```python
def clean_tab(tab,col,val):
    tab.drop(tab[tab[col]==val].index,inplace = True)
def tts_split(X,y,size):
    rs = ShuffleSplit(n_splits=5, test_size=size)
    rs.get_n_splits(X)
    for train_index, test_index in rs.split(X, y): 
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
    return X_train, X_test,y_train, y_test

n_folds = 5

def rmsle_cv(model,X_train,y_train):
    kf = KFold(3, shuffle=True, random_state=None).get_n_splits(X_train)
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train.ravel(), scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
def rmse_ann(y_true, y_pred):
    return ks.backend.sqrt(ks.backend.mean(ks.backend.square(y_pred - y_true), axis=-1))
def rmse_ann2(y_true, y_pred):
    return ks.backend.sqrt(ks.backend.mean(ks.backend.square(y_pred - y_true), axis=-1))
def rmse_ann3(y_true, y_pred):
    return ks.backend.sqrt(ks.backend.mean(ks.backend.square(y_pred - y_true), axis=-1))
def plot_history(history):
    hist = pd.DataFrame(history.history)
    hist['epoch'] = history.epoch

    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Abs Error [MPG]')
    plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
    plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
    plt.ylim([0,1])
    plt.legend()

    plt.figure()
    plt.xlabel('Epoch')
    plt.ylabel('Mean Square Error [$MPG^2$]')
    plt.plot(hist['epoch'], hist['mse'],
               label='Train Error')
    plt.plot(hist['epoch'], hist['val_mse'],
               label = 'Val Error')
    plt.ylim([0,1])
    plt.legend()
    plt.show()
# Mostra o progresso do treinamento imprimindo um único ponto para cada epoch completada
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

def get_features_targets(data):
    features = np.zeros(shape=(len(data), 4))
    
    features[:, 0] = data['MAG_AUTO_G'] - data['MAG_AUTO_R']
    features[:, 1] = data['MAG_AUTO_R'] - data['MAG_AUTO_I']
    features[:, 2] = data['MAG_AUTO_I'] - data['MAG_AUTO_Z']
    features[:, 3] = data['MAG_AUTO_Z'] - data['MAG_AUTO_Y']
    #features[:, 4] = data['WAVG_MAG_PSF_G'] - data['WAVG_MAG_PSF_R']
    #features[:, 5] = data['WAVG_MAG_PSF_R'] - data['WAVG_MAG_PSF_I']
    #features[:, 6] = data['WAVG_MAG_PSF_I'] - data['WAVG_MAG_PSF_Z']
    #features[:, 7] = data['WAVG_MAG_PSF_Z'] - data['WAVG_MAG_PSF_Y']

    
    
    targets = data['z']
    return features, targets
```

## Machine Learning

```python
X,y = get_features_targets(vvds)
```

```python
X.shape
```

```python
X = np.concatenate((X,vvds["MAG_AUTO_I"].values.reshape(-1,1)),axis = 1)
```

```python
X.shape
```

```python
X_train, X_test,y_train, y_test = tts_split(X,y.values,0.3)
```

```python
EarlyStop = EarlyStopping(monitor='rmse_ann', mode='min', patience=15)
n_inputs = X_train.shape[1]
```

```python
ann_model = Sequential([Dense(n_inputs,input_shape = X_train.shape[1:],kernel_initializer='normal', kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)),
                       Dropout(0.05),
                       Dense(10, kernel_initializer='normal', activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)),
                       Dropout(0.45),
                       Dense(10,kernel_initializer='normal', activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)) ,
                       Dense(1,activation = None,name = "output")
                       ])
opt = ks.optimizers.Adam(lr=0.001, decay=5e-4)
#opt = tf.keras.optimizers.RMSprop(0.001)
ann_model.compile(optimizer=opt, loss=rmse_ann, metrics=['mse', 'mae', 'mape',rmse_ann])
```

```python
history = ann_model.fit(X_train, y_train, epochs=512,batch_size=64, validation_split=0.2, callbacks=[EarlyStop])
```

```python
plot_history(history)
```

```python
loss, mae, mse,mape,rmse_ann = ann_model.evaluate(X_test, y_test)

print("Testing set Mean Abs Error: {:5.2f} ".format(mae))
print("\n")
print("Testing set Root Mean Abs Error: {:5.2f} ".format(rmse_ann))
```

## Stacked method

```python
def build_nn():

    ann_model = Sequential([Dense(n_inputs,input_shape = X_train.shape[1:],kernel_initializer='normal', activation='relu', kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)),
                       Dropout(0.05),
                       Dense(10, kernel_initializer='normal', activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)),
                       Dropout(0.45),
                       Dense(10,kernel_initializer='normal', activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=1e-5, l2=1e-4),
                              bias_regularizer=regularizers.l2(1e-4),activity_regularizer=regularizers.l2(1e-5)) ,
                       Dense(1,activation = None,name = "output")
                       ])
    opt = ks.optimizers.Adam(lr=0.001, decay=5e-4)
    #opt = tf.keras.optimizers.RMSprop(0.001)
    ann_model.compile(optimizer=opt, loss=rmse_ann2, metrics=['mse', 'mae', 'mape',rmse_ann2])
    
    return ann_model
```

```python
DTR_boost = AdaBoostRegressor(base_estimator= DecisionTreeRegressor(max_depth=10, min_samples_split=30, splitter='random'),
                             learning_rate=0.01, loss='exponential', n_estimators=200)
DTR_bag = BaggingRegressor(base_estimator=DecisionTreeRegressor(max_depth=10, min_samples_split=30, splitter='random'),n_estimators=200)
```

```python
models = [AdaBoostRegressor(base_estimator=GradientBoostingRegressor(criterion='mae',
                                                            loss='lad',
                                                            max_features='log2',
                                                            n_estimators=200),
                   learning_rate=0.01, loss='exponential', n_estimators=200),
 BaggingRegressor(base_estimator=RandomForestRegressor(criterion='mae',
                                                       max_depth=13,
                                                       min_samples_split=20),
                  n_estimators=200, n_jobs=-1),DTR_boost,DTR_bag]
```

```python
ann1_clf = tf.keras.wrappers.scikit_learn.KerasRegressor(
                            build_nn,
                            epochs=200,batch_size=64,validation_split=0.2,
                            verbose=False)
```

```python
ann1_clf._estimator_type = 'regressor'
```

```python
models.append(ann1_clf)
models
```

```python
stacked = StackingRegressor(estimators=[('adaGbost', models[0]),
                         ('ann', ann1_clf),
                         ('dtr',models[2])] ,final_estimator=models[1],cv = 3)
```
